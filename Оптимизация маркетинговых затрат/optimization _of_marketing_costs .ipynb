## Описание проекта   
Необходимо помочь маркетологамЯндекс.Афиши оптимизировать маркетинговые затраты.  
У вас в распоряжении есть данные с июня 2017 по конец мая 2018 года:  
- лог сервера с данными о посещениях сайта Яндекс.Афиши,
- выгрузка всех заказов за этот период,
- статистика рекламных расходов.

<ins>***Описание данных***</ins>

Таблица **visits** (лог сервера с информацией о посещениях сайта):
- **Uid** — уникальный идентификатор пользователя
- **Device** — категория устройства пользователя
- **Start Ts** — дата и время начала сессии
- **End Ts** — дата и время окончания сессии
- **Source Id** — идентификатор рекламного источника, из которого пришел пользователь  

Таблица **orders** (информация о заказах):
- **Uid** — уникальный id пользователя, который сделал заказ
- **Buy Ts** — дата и время заказа
- **Revenue** — выручка Яндекс.Афиши с этого заказа  

Таблица **costs** (информация о затратах на маркетинг):
- **source_id** — идентификатор рекламного источника
- **dt** — дата
- **costs** — затраты на этот рекламный источник в этот день

**Цель:**  
В ходе анализа мы попытаемся выяснить, как обстоят дела с продажами на нашем сайте, сколько нам удалось привлечь новых посетителей и как они себя вели на протяжении исследуемого периода. Эта нам поможет в дальнейшем для выстраивания политики компании по привлечению и удержанию своих клиентов. Кроме этого проанализируем затраты на рекламу и ее отдачу, выясним, оптимально ли мы тратим деньги на маркетинг.

## Подготовка отчета и построение метрик 

### Исследование продукта

#### Сколько людей пользуются в день, неделю, месяц

Чтобы посчитать дневное, недельное и месячное использование, выделим в отдельные столбцы год, месяц, неделю, а также полную дату:

```python
visits_copy = visits.copy()
```

```python
visits_copy['session_year'] = visits_copy['session_start'].dt.year
visits_copy['session_week']  = visits_copy['session_start'].dt.week
visits_copy['session_month']  = visits_copy['session_start'].dt.month
visits_copy['session_date'] = visits_copy['session_start'].dt.date
```

Посчитаем метрики. Сгруппируем данные по уникальным пользователям и найдём среднее:

```python
dau_total = visits_copy.groupby('session_date').agg({'user_id': 'nunique'})
wau_total = visits_copy.groupby(['session_year', 'session_week']).agg({'user_id': 'nunique'})
mau_total = visits_copy.groupby(['session_year', 'session_month']).agg({'user_id': 'nunique'})

print('Среднее колличество пользователей в день:', int(dau_total.mean()))
print('Среднее колличество пользователей в неделю:', int(wau_total.mean()))
print('Среднее колличество пользователей в месяц:', int(mau_total.mean()))
```

Для наглядности отразим на графиках динамику данных показателей:

```python
f, axes = plt.subplots(nrows=3, ncols=1, figsize=(15,15))
plt.subplots_adjust(hspace=0.3)

dau_total.plot(ax=axes[0])
axes[0].set_title('Динамика DAU')
axes[0].set_xlabel('Дата')
axes[0].set_ylabel('Количество пользователей')
axes[0].get_legend().remove()

wau_total.plot(ax=axes[1])
axes[1].set_title('Динамика WAU')
axes[1].set_xlabel('Дата')
axes[1].set_ylabel('Количество пользователей')
axes[1].get_legend().remove()

mau_total.plot(ax=axes[2])
axes[2].set_title('Динамика MAU')
axes[2].set_xlabel('Дата')
axes[2].set_ylabel('Количество пользователей')
axes[2].get_legend().remove()
```

Посмотрим, как эти метрики различаются в зависимости от устройств пользователей и рекламного источника:

```python
f, axes = plt.subplots(nrows=3, ncols=1, figsize=(15,15))
plt.subplots_adjust(hspace=0.3)

visits_copy.groupby(['session_date', 'device']).agg({'user_id': 'nunique'}).unstack().plot(ax=axes[0])
axes[0].set_title('Динамика DAU в зависимости от устройства пользователей')
axes[0].set_xlabel('Дата')
axes[0].set_ylabel('Количество пользователей')
axes[0].legend(('desktop', 'touch'), title='Устройство пользователя')

visits_copy.groupby(['session_year', 'session_week', 'device']).agg({'user_id': 'nunique'}).unstack().plot(ax=axes[1])
axes[1].set_title('Динамика WAU в зависимости от устройства пользователей')
axes[1].set_xlabel('Дата')
axes[1].set_ylabel('Количество пользователей')
axes[1].legend(('desktop', 'touch'), title='Устройство пользователя')

visits_copy.groupby(['session_year', 'session_month', 'device']).agg({'user_id': 'nunique'}).unstack().plot(ax=axes[2])
axes[2].set_title('Динамика MAU в зависимости от устройства пользователей')
axes[2].set_xlabel('Дата')
axes[2].set_ylabel('Количество пользователей')
axes[2].legend(('desktop', 'touch'), title='Устройство пользователя');
```

```python
f, axes = plt.subplots(nrows=3, ncols=1, figsize=(18,25))
plt.subplots_adjust(hspace=0.3)

visits_copy.groupby(['session_date', 'source_id']).agg({'user_id': 'nunique'}).unstack().plot(ax=axes[0])
axes[0].set_title('Динамика DAU в зависимости от рекламного источника')
axes[0].set_xlabel('Дата')
axes[0].set_ylabel('Количество пользователей')
axes[0].legend(('1','2','3','4','5','6','7','9','10'),title='Рекламный источник')

visits_copy.groupby(['session_year', 'session_week', 'source_id']).agg({'user_id': 'nunique'}).unstack().plot(ax=axes[1])
axes[1].set_title('Динамика WAU в зависимости от рекламного источника')
axes[1].set_xlabel('Дата')
axes[1].set_ylabel('Количество пользователей')
axes[1].legend(('1','2','3','4','5','6','7','9','10'),title='Рекламный источник')

visits_copy.groupby(['session_year', 'session_month', 'source_id']).agg({'user_id': 'nunique'}).unstack().plot(ax=axes[2])
axes[2].set_title('Динамика MAU в зависимости от рекламного источника')
axes[2].set_xlabel('Дата')
axes[2].set_ylabel('Количество пользователей')
axes[2].legend(('1','2','3','4','5','6','7','9','10'),title='Рекламный источник');
```
